```python
import os
from dotenv import load_dotenv
import json
from openai import OpenAI
from IPython.display import Markdown, display

load_dotenv(override=True)
```

```python
openai_api_key = os.getenv("OPENAI_API_KEY")
google_api_key = os.getenv("GOOGLE_API_KEY")
groq_api_key =  os.getenv("GROQ_API_KEY")

if openai_api_key:
    print("Api key de openai esta OK. Empieza con ", openai_api_key[0:3])

if google_api_key:
    print("Api key de google esta OK. Empieza con ", google_api_key[0:3])

if groq_api_key:
    print("Api key de groq esta OK. Empieza con ", groq_api_key[0:3])
```

```python
answers=[]
competitors=[]
```

```python

def procesarMensaje(modelo, mensaje, apiKey=None, baseUrl=None):
    openai= OpenAI(api_key=apiKey, base_url= baseUrl)
    response = openai.chat.completions.create(
        model=modelo,
        messages=mensaje
    )
    answer = response.choices[0].message.content
    answers.append(answer)
    competitors.append(modelo)
    return answer
```

```python
request = "Por favor, propon una pregunta compleja y con matices que pueda plantear a varios LLM para identificar cual es la mejor idea para desarrollar un proyecto con IA en la industria de videojuegos."
request += "Responde solo con la pregunta, sin explicaciones."
messages = [{"role":"user", "content":request}]
```

```python
model = "gpt-4o-mini"
question = procesarMensaje(model, messages)
print(question)
```

```python
messages = [{"role":"user", "content":question}]
openaiMessage = procesarMensaje("gpt-4o-mini",messages)
googleMessage = procesarMensaje("gemini-2.5-flash",messages, apiKey=google_api_key, baseUrl="https://generativelanguage.googleapis.com/v1beta/openai/")
groqMessages = procesarMensaje("llama-3.3-70b-versatile", messages, apiKey=groq_api_key, baseUrl="https://api.groq.com/openai/v1/")
```

```python
prompt_openai_to_gemini = f"""
Has leido las respuestas de 1 modelo de IA a una pregunta {question}.
respuesta del primer competidor: {answers[1]}
Como experto en IA, critícala desde un punto de vista experto. ¿Qué puntos son débiles, discutibles o poco fundamentados? en esta respuesta."""

prompt_openai_to_groq = f"""
Has leido las respuestas de 1 modelo de IA a una pregunta {question}.
Respuesta del segundo competidor: {answers[2]}
Como experto en IA, critícala desde un punto de vista experto. ¿Qué puntos son débiles, discutibles o poco fundamentados? en esta respuesta."""

prompt_gemini_to_openai = f"""
Has leido las respuestas de 1 modelo de IA a una pregunta {question}.
respuesta del primer competidor: {answers[0]}
Como experto en IA, critícala desde un punto de vista experto. ¿Qué puntos son débiles, discutibles o poco fundamentados? en esta respuesta."""

prompt_gemini_to_groq = f"""
Has leido las respuestas de 1 modelo de IA a una pregunta {question}.
Respuesta del primer competidor: {answers[2]}
Como experto en IA, critícala desde un punto de vista experto. ¿Qué puntos son débiles, discutibles o poco fundamentados? en esta respuesta."""

prompt_groq_to_openai = f"""
Has leido las respuestas de 1 modelo de IA a una pregunta {question}.
Respuesta del primer competidor: {answers[0]}
Como experto en IA, critícala desde un punto de vista experto. ¿Qué puntos son débiles, discutibles o poco fundamentados? en esta respuesta."""

prompt_groq_to_gemini = f"""
Has leido las respuestas de 1 modelo de IA a una pregunta {question}.
Respuesta del primer competidor: {answers[1]}
Como experto en IA, critícala desde un punto de vista experto. ¿Qué puntos son débiles, discutibles o poco fundamentados? en esta respuesta."""
```

```python
messages = [{"role":"user", "content":prompt_openai_to_gemini}]
answer_openai_to_gemini = procesarMensaje("gpt-4o-mini",messages)

messages = [{"role":"user", "content":prompt_openai_to_groq}]
answer_openai_to_groq = procesarMensaje("gpt-4o-mini",messages)
```
